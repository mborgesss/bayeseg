```python
from tqdm import tqdm

s1 = 1.
ls2 = np.unique(np.concatenate([np.linspace(0.75, 1, 10), np.linspace(1, 1.25, 10)]))


mciter = 20000
mcburn = 20000
nchains = 1

N = [1000, 10000, 100000]


ss = SeqSeg(replicate = False)
alpha = 0.05
ss.initialize(1, alpha, mciter, mcburn, nchains)

ntests = 10

res = []
for n in N:
    for s2 in ls2:
        #file = filelist[2]
        #fs, wave = op.readFile(file)
        mev = 0.
        n2 = int(n/2)

        for i in range(ntests):
            wave = np.concatenate([np.random.normal(0, s1, n2), np.random.normal(0, s2, n2)])
            ss.initialize(1, alpha, mciter, mcburn, nchains)

            ss.feed_data(wave)            
            ev = ss.tester(n2, 0)
            #print(ev)
            mev = mev + ev/ntests

        res.append([n, s2, mev])
        #print("N = {}, s2 = {:.2}, ev = {:.4}".format(n, s2, mev))
```

```python
dftest = pd.DataFrame.from_records(res, columns = ['N' , 's2', 'ev'])
```

```python
fig = plt.figure(figsize = (20, 10))
from itertools import cycle
lines = ["-","--","-.",":"]
linecycler = cycle(lines)
plt.rcParams['font.size'] = 32
for N in dftest[dftest.N > 500].N.unique():
    dftmp = dftest[dftest.N == N]
    plt.plot(dftmp.s2, dftmp.ev, 'k-o', label = "N = {}".format(N), linestyle = next(linecycler))

    
plt.legend(fontsize = 32, loc = 'best')
plt.grid(True)
#plt.savefig('/home/paulo/Documents/Papers/Changepoint/figuras/evsim.pdf', dpi = 1600, bbox_inches = 'tight')
plt.show()
```

```python
from tqdm import tqdm

s1 = 1.
ls2 = np.unique(np.concatenate([np.linspace(0.75, 1, 10), np.linspace(1, 1.25, 10)]))


mciter = 20000
mcburn = 20000
nchains = 1

N = [1000, 10000, 100000]


ss = SeqSeg(replicate = False)
alpha = 0.05
ss.initialize(1, alpha, mciter, mcburn, nchains)

ntests = 100

res = []
for n in N:
    for s2 in ls2:
        #file = filelist[2]
        #fs, wave = op.readFile(file)
        mev = 0.
        n2 = int(n/2)
        power = 0.
        
        for i in range(ntests):
            np.random.seed()            
            
            wave = np.concatenate([np.random.normal(0, s1, n2), np.random.normal(0, s2, n2)])
            
            ss.initialize(1, alpha, mciter, mcburn, nchains)            
            ss.feed_data(wave)
            ev = ss.tester(n2, 0, regularize = True)
            print("N = {}, s2 = {:.2}, ev = {:.4}".format(n, s2, ev))
            mev = mev + ev/ntests
            if ev < alpha:
                power = power + 1/ntests

        res.append([n, s2, mev, power])
        #print("N = {}, s2 = {:.2}, ev = {:.4}".format(n, s2, mev))
```

```python
dftest = pd.DataFrame.from_records(res, columns = ['N' , 's2', 'ev', 'power'])
```

```python
fig = plt.figure(figsize = (20, 10))
from itertools import cycle
lines = ["-","--","-.",":"]
linecycler = cycle(lines)
plt.rcParams['font.size'] = 32
for N in dftest[dftest.N > 500].N.unique():
    dftmp = dftest[dftest.N == N]
    plt.plot(dftmp.s2, dftmp.power, 'k-o', label = "N = {}".format(N), linestyle = next(linecycler))

    
plt.legend(fontsize = 32, loc = 'best')
plt.grid(True)
#plt.savefig('/home/paulo/Documents/Papers/Changepoint/figuras/evpower.pdf', dpi = 1600, bbox_inches = 'tight')
plt.show()
```
